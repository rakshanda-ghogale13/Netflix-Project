## Q.2. Show Top 10 Directors, who gave the highest number of TV Shows & Movies to Netflix ?

### value_counts()

```python
# data['director'].value_counts().head(10)

data.head(2)
```

```python
data['director'].value_counts().head(10)
```

```python
top_directors = data['director'].value_counts().head(10)

# Plotting
plt.figure(figsize=(10, 6))
top_directors.plot(kind='bar', color='skyblue')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Movies and Tv Shows')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
